***Вопрос 4***

При нулевой экономической прибыли прекращать выпуск нецелесообразно, так как внешние условия могут привести к увеличению цены на товар/снижению издержек. Даже при работе компании в убыток иногда имеет смысл продолжать производство, так как убытки, понесённые компанией в этом случае, могут быть ниже, чем постоянные издержки, которые компания вынуждена будет заплатить в любом случае. В июне мы как раз наблюдаем подобную ситуацию. В июле убыток компании при $Q = 2000$ и $Q = 3000$ совпадает с постоянными издержками, поэтому производство можно как продолжать, так и останавливать. А в августе при любых объемах производства убыток компании превышает постоянные издержки, поэтому производство необходимо останавливать.

***Вопрос 5***

```python
import pandas as pd
import numpy as np

data = pd.read_excel('data.xlsx')

plt.plot(data.Q, data.FC)
plt.plot(data.Q, data.VC)
plt.plot(data.Q, data.TC)
plt.xlabel('Тыс. штук')
plt.ylabel('Тыс. евро')
plt.axvline(x=0, c='black')
plt.axhline(y=0, c='black')
plt.legend(('FC', 'VC', 'TC'), loc=9)
plt.show()
```

```python
plt.plot(data.Q, data.TR0)
plt.plot(data.Q, data.TR1)
plt.plot(data.Q, data.TR2)
plt.plot(data.Q, data.TR3)
plt.plot(data.Q, data.TR4)
plt.plot(data.Q, data.TC, linewidth=5)
plt.xlabel('Тыс. штук')
plt.ylabel('Тыс. евро')
plt.axvline(x=0, c='black')
plt.axhline(y=0, c='black')
plt.legend(('Апрель', 'Май', 'Июнь', 'Июль', 'Август', 'TC'), loc=9)
plt.show()
```

Соответственно зоны, в которых график $TR$ проходит над графиком $TC$ являются зонами прибыльности, в которых эти графики совпадают - неубыточности, остальные зоны явдяются зонами убыточности

***Вопрос 6***

```python
plt.plot(data.Q, data.AFC)
plt.plot(data.Q, data.AVC)
plt.plot(data.Q, data.ATC)
plt.plot(data.Q, data.MC)
plt.xlabel('Тыс. штук')
plt.ylabel('Евро')
plt.axvline(x=0, c='black')
plt.axhline(y=0, c='black')
plt.legend(('AFC', 'AVC', 'ATC', 'MC'), loc=9)
plt.show()
```

Оптимальный выпуск продукции фирмы достигается при выполнении равенства $p = MC$ при заданной цене. Интерполируя $MC$ линейно, вычислим точный объем продукции для каждого месяца.

```python
p = [40, 30, 25, 20, 16]
opt_val = np.interp(p, data.MC, data.Q)
plt.plot(data.Q, data.MC, c='black')
for i, p_ in enumerate(p):
    plt.scatter(opt_val[i], p_, linewidths=8)
plt.legend(('MC', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август'))
plt.show()
```
